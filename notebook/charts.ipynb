```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
n_samples = 100

data = {
    'Shot': np.repeat(['0-shot', '2-shot', '5-shot'], n_samples),
    'Scores': np.tile(['Bleu', 'Charf', 'Ter'], n_samples),
    'Value': np.random.rand(3 * n_samples)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate mean values for each metric by 'Shot'
mean_values = df.groupby(['Shot', 'Scores'])['Value'].mean().reset_index()

# Set the style to remove the background grid
sns.set(style='white')  # You can also use 'ticks' for a different style

# Plot the histogram
g = sns.FacetGrid(df, col='Shot', hue='Scores', sharex=False, height=4, aspect=1.2)
g.map(sns.histplot, 'Value', bins=3, alpha=0.7)

# Customize the legend
g.add_legend(title='Scores')

# Remove x-axis ticks and labels
g.set(xticks=[], xticklabels=[])

# Set custom subtitles under each subplot
titles = ["0-shot", "2-shot", "5-shot"]

for ax, title in zip(g.axes.flat, titles):
    ax.set_xlabel(title)

# Remove titles above each subplot
g.set_titles("")

# Add metric values as text annotations
for ax, title in zip(g.axes.flat, titles):
    metrics = mean_values[mean_values['Shot'] == title]
    for i, (_, row) in enumerate(metrics.iterrows()):
        ax.text(0.5, 0.85 - i * 0.15, f"{row['Scores']}: {row['Value']:.2f}", 
                transform=ax.transAxes, fontsize=10, ha='center')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the specific metric values
values = {
    'BLEU': 11.51,
    'CHRF': 37.5,
    'TER': 122.52
}

# Create a DataFrame from the values
df = pd.DataFrame(list(values.items()), columns=['Scores', 'Value'])

# Set the style to remove the background grid and x-axis line
sns.set(style='ticks')  # You can also use 'white' or 'darkgrid' for different styles
sns.set_palette('pastel')

# Create a horizontal bar plot
plt.figure(figsize=(5, 3))  # Set the figure size

# Plot the horizontal bar plot
ax = sns.barplot(data=df, x='Value', y='Scores', errorbar=None, palette=sns.color_palette('pastel'))

# Annotate the bars with score values to the right of the bars
for index, row in df.iterrows():
    plt.text(row['Value'] + 1, index, f'{row["Value"]:.2f}', va='center', fontsize=10)

# Remove the y-axis label ("Scores")
plt.ylabel("")

# Remove x-axis ticks and labels
plt.xticks([])

# Remove the x-axis line (bottom spine)
ax.spines['bottom'].set_visible(False)
# Remove the x-axis line (bottom spine) and the right y-axis line
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Save the plot as an image file (e.g., PNG)
plt.savefig('Google_translate_api.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
